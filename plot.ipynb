```python
import numpy as np
import matplotlib.pyplot as plt
import csv
```

## Profile1

```python
x = []
y1 = []
y2 = []

with open('profile1.data','r') as data:
    lines = data.read().split("\n")    
    r = 0
    time = 0
    min_flt = 0
    maj_flt = 0
    for r in range(len(lines)-2):
        item = lines[r].split(' ')
        if r == 0:
            time -= int(item[0])
        if r == len(lines)-3:
            time += int(item[0])
        min_flt += int(item[1])
        maj_flt += int(item[2])
        y1.append(min_flt)
        y2.append(maj_flt)
        x.append(r)
        r += 0.05 
```

```python
time
```

```python
plt.plot(x,y1,label="minor page fault")
plt.plot(x,y2,label="major page fault")
plt.xlabel('time(s)')
plt.ylabel('Accumulated Page Fault Count')
plt.title('Page Fault Count of Two Random Access')
plt.legend()
plt.show()
```

## Profile 3

```python
def getVal(fname):
    with open(fname,'r') as data:
        lines = data.read().split("\n")  
        time = 0
        tot = 0
        for r in range(len(lines)-2):
            item = lines[r].split(' ')
            if r == 0:
                time -= int(item[0])
            if r == len(lines)-3:
                time += int(item[0])
            tot += int(item[3])
        print(time)
    return tot/time
```

```python
num1 = getVal('profile3-1.data')
num2 = getVal('profile3-5.data')
num3 = getVal('profile3-11.data')
num4 = getVal('profile3-20.data')
num5 = getVal('profile3-22.data')
```

```python
name_list = ['1','5','11','20', '22']
y = [num1, num2, num3, num4, num5]
x = range(len(y))
plt.bar(x, y, facecolor='#9999ff', width = 0.5, edgecolor='white', tick_label=name_list, alpha=0.8)
plt.xlabel("Number of Process")   
plt.ylabel("Average Utilization")  
plt.title("CPU Utilization of Process 5")
for a,b in zip(x,y):
    plt.text(a, b+0.0002, '%.4f' % b, ha='center', va= 'bottom',fontsize=8)
plt.show()
```
